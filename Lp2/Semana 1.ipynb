## **üåü 1. Introducci√≥n a Jupyter Notebook**

## **üìù 2. Markdown y LaTeX en Jupyter**

### **2.1 ¬øQu√© es Markdown?**  
Markdown es un lenguaje de marcado ligero que facilita el formateo de texto de manera sencilla y elegante.

### **2.2 Principales Comandos Markdown**
- **Encabezados:** `# T√≠tulo 1`, `## T√≠tulo 2`, etc. (hasta 6 #)
- **Listas:** 
    - `*`, `-` o `+` para listas desordenadas.
    - `1.`, `2.` para listas ordenadas.
- **Enlaces e im√°genes:** `[Texto](url)` para enlaces, `![Alt](url)` para im√°genes.

### **2.3 Uso de LaTeX para Matem√°ticas**  
LaTeX es ideal para escribir ecuaciones matem√°ticas. Ejemplos:

- **Ecuaciones en l√≠nea:** `$E = mc^2$`.
- **Ecuaciones centradas:** `$$ E = mc^2 $$`.

Ejemplo de una ecuaci√≥n m√°s compleja:

üìà **Matplotlib & Seaborn:** Crear un gr√°fico con estilo

````python
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Seno de x', color='purple', linewidth=2)
plt.title("Gr√°fico de Seno", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("sin(x)", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
````
